```python
from pathlib import Path
import pandas as pd

schema_path = Path("schema.json")
if not schema_path.exists():
  print(f"Schema file not found: {schema_path}")
else:
  with schema_path.open("r", encoding="utf-8") as fh:
      schemas = pd.read_json(fh, orient="records")
      print(schemas)
```

```python
import pandas as pd
import sqlalchemy


engine = sqlalchemy.create_engine("sqlite:///../app.db")

db = engine.connect()

data = db.execute(sqlalchemy.text("SELECT * FROM prices")).fetchall()

print(data[0]._asdict())
for row in data:
    print(row)
```
